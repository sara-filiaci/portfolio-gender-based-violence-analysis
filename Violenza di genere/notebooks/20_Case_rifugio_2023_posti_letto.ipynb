## Note sui dati

I valori nelle colonne rappresentano **percentuali (%)**.  
Ogni riga indica la distribuzione percentuale dei motivi di uscita dalle Case rifugio per regione.

```python
import pandas as pd

df = pd.read_csv("../raw/Case_rifugio_2023_posti_letto.csv")

df.head()
```

```python
# Per ogni colonna: numero di valori unici + elenco valori

for col in df.columns:
    print("=" * 60)
    print(f"Colonna: {col}")
    print(f"Numero valori unici: {df[col].nunique(dropna=False)}")
    print("Valori unici:")
    print(df[col].unique())
    print("\n")
```

```python
# Rimozione righe completamente vuote

df = df.dropna(how="all")

print("Shape dopo rimozione righe vuote:", df.shape)
```

```python
# Standardizzazione nomi colonne

df = df.rename(columns={
    "Regione": "regione",
    "Case rifugio rispondenti": "case_rifugio_rispondenti",
    "Donne ospitate": "donne_ospitate",
    "Numero totale di posti letto autorizzati": "posti_letto_autorizzati_totale",
    "Numero medio di posti letto autorizzati in Casa rifugio": "posti_letto_autorizzati_medio",
    "Numero totale di posti letto effettivamente utilizzati ": "posti_letto_utilizzati_totale",
    "Numero medio di posti letto effettivamente utilizzati in Casa rifugio": "posti_letto_utilizzati_medio",
    "Numero medio di notti per donna": "notti_medie_per_donna"
})

print(df.columns.tolist())
```

```python
# Pulizia colonna regione

df["regione"] = (
    df["regione"]
    .str.lower()
    .str.strip()
    .str.replace("-", " ", regex=False)
)

df["regione"] = df["regione"].replace({
    "valle d'aosta/vallée d'aoste": "valle d'aosta",
    "trentino alto adige/südtirol": "trentino alto adige",
    "bolzano/bozen": "bolzano",
    "nord-ovest": "nord ovest",
    "nord-est": "nord est"
})

print(df["regione"].unique())
```

```python
# Conversione colonne numeriche

cols = df.columns[1:]

for c in cols:
    df[c] = (
        df[c]
        .astype(str)
        .str.replace(",", ".", regex=False)
        .str.strip()
        .replace("-", None)
        .astype(float)
    )
```

```python
# Verifica finale

df.head()
df.info()

print("\nMissing per colonna:")
print(df.isna().sum())

print("\nRighe duplicate:")
print(df.duplicated().sum())
```

```python
# Salvataggio dataset pulito

output_path = "../data_clean/case_rifugio_2023_posti_letto_clean.csv"

df.to_csv(output_path, index=False, encoding="utf-8-sig")

print("File salvato in:", output_path)
```
