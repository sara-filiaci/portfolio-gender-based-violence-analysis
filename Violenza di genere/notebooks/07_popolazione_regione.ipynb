```python
import pandas as pd

df = pd.read_csv(
    "../raw/Popolazione per regione.csv",
    sep=";"
)

df.head()
```

```python
# Standardizzazione automatica nomi colonne:
# - tutto minuscolo

df.columns = df.columns.str.lower()

# Rinomina colonne in modo standardizzato (snake_case, senza accenti/spazi)
df = df.rename(columns={
    "regione": "territorio",
    "uniti civilmente": "uniti_civilmente",
    "maschi già in unione civile (per scioglimento unione)": "maschi_ex_unione_scioglimento",
    "maschi già in unione civile (per decesso del partner)": "maschi_ex_unione_decesso",
    "totale maschi": "totale_maschi",
    "coniugate": "coniugate",
    "unite civilmente": "unite_civilmente",
    "femmine già in unione civile (per scioglimento unione)": "femmine_ex_unione_scioglimento",
    "femmine già in unione civile (per decesso del partner)": "femmine_ex_unione_decesso",
    "totale femmine": "totale_femmine"
})

# Visualizza nomi colonne
df.columns.tolist()
```

```python
# Per ogni colonna: numero di valori unici + elenco valori

for col in df.columns:
    print("=" * 60)
    print(f"Colonna: {col}")
    print(f"Numero valori unici: {df[col].nunique(dropna=False)}")
    print("Valori unici:")
    print(df[col].unique())
    print("\n")
```

```python
# Pulizia colonna territorio:
# - uniformazione minuscolo
# - correzione denominazioni incoerenti / verbose


df["territorio"] = (
    df["territorio"]
    .str.lower()
    .str.strip()
    .str.replace("-", " ", regex=False)
)

df["territorio"] = df["territorio"].replace({
    "valle d'aosta/vallée d'aoste": "valle d'aosta",
    "trentino alto adige/südtirol": "trentino alto adige"
    })

# Verifica tipo delle colonne
df.dtypes
```

```python
# Verifica finale post-pulizia (struttura + qualità base)
display(df.head(10))
df.info()

print("\nMissing values per colonna:")
display(df.isna().sum().sort_values(ascending=False))

print("\nNumero valori unici per colonna:")
display(df.nunique(dropna=False).sort_values(ascending=False))

print("\nRighe duplicate:", df.duplicated().sum())

for col in df.columns:
    print("\nColonna:", col)
    if col == "periodo":
        print(sorted(df[col].unique().tolist()))
    else:
        print(df[col].unique())
```

```python
# Salvataggio dataset pulito

output_path = "../data_clean/popolazione_regione.csv"

df.to_csv(
    output_path,
    index=False,
    encoding="utf-8-sig"
)

print("File salvato in:", output_path)
```
