Questa tabella è relativa solo all'anno 2018 e riporta la composizione percentuale (%) delle risposte di persone di nazionalità italiana e di età compresa fra i 18 e i 74 anni.

```python
import pandas as pd

df = pd.read_csv(
    "../raw/Opinioni_ruoli_tradizionali_reg.csv",
    sep=";",
    encoding="utf-8-sig",
)

df.head()
```

```python
# Per ogni colonna: numero di valori unici + elenco valori

for col in df.columns:
    print("=" * 60)
    print(f"Colonna: {col}")
    print(f"Numero valori unici: {df[col].nunique(dropna=False)}")
    print("Valori unici:")
    print(df[col].unique())
    print("\n")
```

```python
# Rimozione colonne non informative o ridondanti:
# - colonne costanti (un solo valore per tutto il dataset)
# - colonne duplicate sul piano informativo
# - colonne completamente vuote

cols_to_drop = [
    "FREQ",               
    "Frequenza",          
    "REF_AREA",          
    "DATA_TYPE",          
    "Indicatore",        
    "SEX",                     
    "STEREOT_GENDROLEPARTNER",         
    "DEGREE_AGREEMENT" ,
    "TIME_PERIOD",
    "OBS_STATUS"
]

df = df.drop(columns=cols_to_drop)

# Verifica struttura finale
print("Nuova shape:", df.shape)
print("Colonne rimanenti:", df.columns.tolist())
```

```python
# Standardizzazione nomi colonne:
# - tutto minuscolo
# - nomi più chiari e coerenti

df = df.rename(columns={
    "Territorio": "territorio",
    "Sesso": "sesso",
    "Stereotipi sui ruoli di genere, comportamenti nella coppia": "imput_affermazione",
    "Grado di accordo": "grado_accordo",
    "Osservazione": "percentuale"
})


# Eliminazione delle righe completamente vuote 
df = df.dropna(how="all")

# Stampa delle colonne e dei valori unici che contengono
for col in df.columns:
    print("\nColonna:", col)
    print(df[col].unique())
```

```python
# Pulizia valori categorici:
# - rimozione voci aggregate "Totale" da tutte le colonne testuali
# - normalizzazione label (sesso, territorio, grado_accordo)
# - pulizia quote / caratteri sporchi
# - conversione percentuale a numerico


# Rimozione della voce aggregata "Totale" (ovunque) per evitare doppio conteggio
cat_cols = df.select_dtypes(include="object").columns

mask_totale = False
for c in cat_cols:
    mask_totale = mask_totale | (
        df[c].astype(str).str.strip().str.lower().eq("totale")
    )

print("Righe rimosse (contengono 'Totale'):", int(mask_totale.sum()))
df = df.loc[~mask_totale].copy()


# 1) sesso: minuscolo + standard
df["sesso"] = df["sesso"].str.lower().str.strip()

df["sesso"] = df["sesso"].replace({
    "maschi": "maschi",
    "femmine": "femmine"
})


# 2) territorio: pulizia caratteri sporchi + minuscolo + correzione denominazioni
df["territorio"] = df["territorio"].str.strip()

df["territorio"] = df["territorio"].str.replace(r"^'+|'+$", "", regex=True)

df["territorio"] = df["territorio"].str.replace('"', "", regex=False)

df["territorio"] = df["territorio"].str.replace("\\", "", regex=False)

df["territorio"] = df["territorio"].str.replace("-", " ", regex=False)

df["territorio"] = df["territorio"].str.lower().str.strip()

df["territorio"] = df["territorio"].replace({
    "trentino alto adige / südtirol": "trentino alto adige",
    "provincia autonoma bolzano / bozen": "bolzano",
    "provincia autonoma trento": "trento",
    "valle d'aosta / vallée d'aoste": "valle d'aosta"
})


# 3) grado_accordo: rimozione quote sporche + normalizzazione
df["grado_accordo"] = df["grado_accordo"].str.strip()

df["grado_accordo"] = df["grado_accordo"].str.replace(r"^'+|'+$", "", regex=True)

df["grado_accordo"] = df["grado_accordo"].str.replace('d"\'accordo', "d'accordo", regex=False)

df["grado_accordo"] = df["grado_accordo"].str.replace('d"accordo', "d'accordo", regex=False)

df["grado_accordo"] = df["grado_accordo"].str.replace('"', "", regex=False)

df["grado_accordo"] = df["grado_accordo"].str.replace("\\", "", regex=False)

df["grado_accordo"] = df["grado_accordo"].str.lower().str.strip()

df["grado_accordo"] = df["grado_accordo"].replace({
    "molto d'accordo": "molto_d_accordo",
    "abbastanza d'accordo": "abbastanza_d_accordo",
    "poco d'accordo": "poco_d_accordo",
    "per niente d'accordo": "per_niente_d_accordo",
    "non risponde": "non_risponde"
})


# 4) imput_affermazione: pulizia testi
df["imput_affermazione"] = df["imput_affermazione"].str.strip()

df["imput_affermazione"] = df["imput_affermazione"].str.replace(r"^'+|'+$", "", regex=True)

df["imput_affermazione"] = df["imput_affermazione"].str.replace('l"\'uomo', "l'uomo", regex=False)

df["imput_affermazione"] = df["imput_affermazione"].str.replace('L"\'uomo', "L'uomo", regex=False)

df["imput_affermazione"] = df["imput_affermazione"].str.replace('"', "", regex=False)

df["imput_affermazione"] = df["imput_affermazione"].str.replace("\\", "", regex=False)

df["imput_affermazione"] = df["imput_affermazione"].str.replace(r"\s*,\s*", ", ", regex=True)

df["imput_affermazione"] = df["imput_affermazione"].str.replace(r"\s+", " ", regex=True)

df["imput_affermazione"] = df["imput_affermazione"].str.strip()


# 5) percentuale: conversione a numerico
df["percentuale"] = pd.to_numeric(df["percentuale"], errors="coerce")


# Verifica rapida post-pulizia
print("\nNuova shape:", df.shape)

for col in ["territorio", "sesso", "imput_affermazione", "grado_accordo"]:
    print("\nColonna:", col)
    print(df[col].unique())
```

```python

# Verifica finale post-pulizia (struttura + qualità base)

# 1) Dimensioni e info colonne
display(df.head(10))
df.info()

# 2) Tipi e missing
print("\nMissing values per colonna:")
display(df.isna().sum().sort_values(ascending=False))

# 3) Valori unici (conteggio) per colonna
print("\nNumero valori unici per colonna:")
display(df.nunique(dropna=False).sort_values(ascending=False))

# 4) Check duplicati (righe identiche)
print("\nRighe duplicate (identiche su tutte le colonne):", df.duplicated().sum())

# 5) Valori distinti per ogni colonna
for col in df.columns:
    print("\nColonna:", col)
    
    if col == "periodo":
        print(sorted(df[col].astype(int).unique().tolist()))
    else:
        print(df[col].unique())
```
