```python
import pandas as pd

df = pd.read_csv("../raw/Case_rifugio_2023_quantità_donne_accolte.csv")

df.head()
```

```python
# Per ogni colonna: numero di valori unici + elenco valori

for col in df.columns:
    print("=" * 60)
    print(f"Colonna: {col}")
    print(f"Numero valori unici: {df[col].nunique(dropna=False)}")
    print("Valori unici:")
    print(df[col].unique())
    print("\n")
```

```python
# Standardizzazione nomi colonne

df = df.rename(columns={
    "Regione": "regione",
    "Totale delle donne accolte durante l'anno": "totale_donne_accolte",
    "quantità donne con figli": "di_cui_con_figli",
    "quantità donne straniere": "di_cui_straniere"
})

print(df.columns.tolist())
```

```python
# Pulizia colonna regione

df["regione"] = (
    df["regione"]
    .str.lower()
    .str.strip()
    .str.replace("-", " ", regex=False)
)

df["regione"] = df["regione"].replace({
    "valle d'aosta/vallée d'aoste": "valle d'aosta",
    "trentino-alto adige/südtirol": "trentino alto adige",
    "trentino alto adige/südtirol": "trentino alto adige",
    "bolzano/bozen": "bolzano",
    "nord-ovest": "nord ovest",
    "nord-est": "nord est"
})

print(df["regione"].unique())
```

```python
# Verifica finale

df.head()
df.info()

print("\nMissing per colonna:")
print(df.isna().sum())

print("\nRighe duplicate:")
print(df.duplicated().sum())
```

```python
# Salvataggio dataset pulito

output_path = "../data_clean/case_rifugio_2023_quantita_donne_accolte_clean.csv"

df.to_csv(output_path, index=False, encoding="utf-8-sig")

print("File salvato in:", output_path)
```
