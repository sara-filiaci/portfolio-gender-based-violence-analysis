## Note sui dati

I valori nelle colonne rappresentano **percentuali (%)**.  
Ogni riga indica la distribuzione percentuale dei motivi di uscita dalle Case rifugio per regione.

```python
import pandas as pd

df = pd.read_csv("../raw/Case_rifugio_2023_motivo_uscita.csv")

df.head()
```

```python
# Per ogni colonna: numero di valori unici + elenco valori

for col in df.columns:
    print("=" * 60)
    print(f"Colonna: {col}")
    print(f"Numero valori unici: {df[col].nunique(dropna=False)}")
    print("Valori unici:")
    print(df[col].unique())
    print("\n")
```

```python
# Standardizzazione nomi colonne

df = df.rename(columns={
    "Regione": "regione",
    "Conclusione percorso uscita violenza": "conclusione_percorso",
    "Abbandono o ritorno dal maltrattante": "abbandono_o_ritorno_dal_maltrattante",
    "Trasferimento ad altre strutture o abitazioni private o raggiungimento limite massimo giorni di permanenza ed altri motivi": "trasferimento_o_altri_motivi"
})

print(df.columns.tolist())
```

```python
# Pulizia colonna regione

df["regione"] = (
    df["regione"]
    .str.lower()
    .str.strip()
)

print(df["regione"].unique())
```

```python
# Conversione colonne percentuali in float

cols = df.columns[1:]

for c in cols:
    df[c] = (
        df[c]
        .astype(str)
        .str.replace(",", ".", regex=False)
        .str.strip()
        .replace("-", None)
        .astype(float)
    )
```

```python
# Verifica finale

df.head()
df.info()

print("\nMissing per colonna:")
print(df.isna().sum())

print("\nRighe duplicate:")
print(df.duplicated().sum())
```

```python
# Salvataggio dataset pulito

output_path = "../data_clean/case_rifugio_2023_motivo_uscita_clean.csv"

df.to_csv(output_path, index=False, encoding="utf-8-sig")

print("File salvato in:", output_path)
```
